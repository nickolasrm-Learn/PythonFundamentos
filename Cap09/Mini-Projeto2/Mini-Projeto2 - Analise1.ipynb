## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
df
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
# Aumenta o tamanho do plot
plt.figure(figsize=(10,7))
# Cria um histograma contando os anos de registro e plota a linha KDE
sns.histplot(data=df, x='yearOfRegistration', kde=True)
# Altera os rótulos do gráfico
plt.xlabel('Ano de registro')
plt.ylabel('Numero de carros')
plt.title('Histograma de ano de registro de carros')
# Obtem a figura no plt
fig = plt.gcf()
# Mostra a figura
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/vehicle-distribution.png")
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
# Aumenta o tamanho do plot
plt.figure(figsize=(10,7))
# Cria um plot de caixas sobre a variação de preços por tipo de veículo
sns.boxplot(data=df, x='vehicleType', y='price')
# Altera os rótulos do gráfico
plt.xlabel('Tipo do veículo')
plt.ylabel('Preço')
plt.title('Variação de preço por tipo do veículo')
# Obtem a figura do plt
fig = plt.gcf()
# Exibe a figura
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/price-vehicleType-boxplot.png")
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
# Aumenta o tamanho do plot
plt.figure(figsize=(10,7))
# Cria um gráfico de barra com a contagem de veículos por tipo do veículo
sns.countplot(data=df, x='vehicleType')
# Altera os rótulos do gráfico
plt.xlabel('Tipo do veículo')
plt.ylabel('Qtde')
plt.title('Quantidade de veículos por tipo de veículo')
# Obtem a figura do plt
g = plt.gcf()
# Exibe a figura
plt.show()
```

```python
# Salvando o plot
g.savefig("plots/Analise1/count-vehicleType.png")
```
