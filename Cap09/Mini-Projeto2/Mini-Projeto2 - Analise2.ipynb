```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
# Aumenta o tamanho do plot
plt.figure(figsize=(10,7))
# Cria um gráfico de barra horizontal com a quantidade de veículos por marca
sns.countplot(data=df, y='brand')
# Altera os rótulos do gráfico
plt.xlabel('Qtde')
plt.ylabel('Marca')
plt.title('Quantidade de veículos por marca')
# Obtem a figura no plt
g = plt.gcf()
# Mostra a figura
plt.show()
```

```python
# Salvando o plot
g.savefig(("plots/Analise2/brand-vehicleCount.png"))
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
# Aumenta o tamanho do plot
plt.figure(figsize=(10,7))
# Cria um gráfico de barra com a quantidade de veículos por tipo de veículo agrupados por câmbio
sns.countplot(data=df, x='vehicleType', hue='gearbox')
# Altera os rótulos do gráfico
plt.xlabel('Tipo do veículo')
plt.ylabel('Qtde')
plt.title('Quantidade de veículos por tipo e câmbio')
# Obtem a figura no plt
fig = plt.gcf()
# Mostra a figura
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise2/vehicletype-gearbox-price.png")
```
