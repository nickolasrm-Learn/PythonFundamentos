## Análise 3

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
#plt.style.use('dark_background')
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
df.info()
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
# Aumenta o tamanho do plot
plt.figure(figsize=(10,7))
# Agrupa os dados por tipo de combustível e câmbio
group_fuel_gearbox = df.groupby(['fuelType', 'gearbox'])
# Cria um data frame da média de preços com grupo anterior e depois desagrupa
df_fuel_gearbox_price_mean = group_fuel_gearbox[['price']].mean().reset_index()
# Cria um grafico de barras com a média de preços pelo tipo de combustível agrupado em câmbio
sns.barplot(data=df_fuel_gearbox_price_mean, x='fuelType', y='price', hue='gearbox')
# Altera os rótulos do gráfico
plt.xlabel('Tipo de combustível')
plt.ylabel('Média de preço')
plt.title('Preço médio por tipo de combustível e câmbio')
# Obtem a figura no plt
fig = plt.gcf()
# Mostra a figura
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
# Aumenta o tamanho do plot
plt.figure(figsize=(10,7))
# Cria um grupo por tipo de veículos e câmbio
group_vehicle_type_gearbox = df.groupby(['vehicleType', 'gearbox'])
# Cria um dataframe com a média de potência dos carros com o grupo anterior e desagrupa o resultado
df_vehicle_type_gearbox_power_mean = group_vehicle_type_gearbox[['powerPS']].mean().reset_index()
# Cria um grafico de barras com o dataframe anterior pelo tipo de veículo agrupado pelo câmbio
sns.barplot(data=df_vehicle_type_gearbox_power_mean, x='vehicleType', y='powerPS', hue='gearbox')
# Altera os rótulos do gráfico
plt.xlabel('Tipo de veículo')
plt.ylabel('Média de potência (PS)')
plt.title('Potência média por tipo de veículo e câmbio')
# Obtem a figura no plt
fig = plt.gcf()
# Mostra a figura
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/vehicletype-fueltype-power.png")
```
