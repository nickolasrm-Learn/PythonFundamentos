```python
# Imports
import pandas as pd
import numpy as np
```

```python
# Carrega o arquivo
load_file = "dados_compras.json"
purchase_file = pd.read_json(load_file, orient = "records")
purchase_file.head()
```

## Informações Sobre os Consumidores

```python
qtde_usuarios = purchase_file['Login'].nunique()
pd.DataFrame({'Qtde usuários': [qtde_usuarios]})
# Implemente aqui sua solução
```

## Análise Geral de Compras

```python
qtde_vendas_unicas = purchase_file['Nome do Item'].nunique()
media_valores = purchase_file['Valor'].mean()
qtde_vendas = purchase_file.size
soma_valores = purchase_file['Valor'].sum()

pd.DataFrame({'Qtde vendas únicas': [qtde_vendas_unicas],
                  'Média preços': media_valores,
                  'Qtde vendas': qtde_vendas,
                  'Rendimento': soma_valores})
# Implemente aqui sua solução
```

## Análise Demográfica

```python
purchase_file[['Login', 'Sexo', 'Idade']]
# Implemente aqui sua solução
```

## Informações Demográficas Por Gênero

```python
qtde_login_por_sexo = purchase_file['Sexo'].value_counts(normalize=True)
pd.DataFrame({'Qtde': np.ceil(qtde_login_por_sexo * qtde_usuarios), 
              'Porcentagem': qtde_login_por_sexo * 100})
# Implemente aqui sua solução
```

## Análise de Compras Por Gênero

```python
grupo_genero = purchase_file.groupby('Sexo')
qtde_vendas_por_genero = grupo_genero.size()
preco_medio_por_genero = grupo_genero['Valor'].mean()
rendimento_por_genero = grupo_genero['Valor'].sum()

display(pd.DataFrame({'Qtde vendas': qtde_vendas_por_genero,
                     'Preço médio': preco_medio_por_genero,
                     'Rendimento': rendimento_por_genero}))

grupo_idade_genero = purchase_file[['Sexo', 'Idade']].groupby(['Sexo', 'Idade'])

pd.DataFrame({'Qtde vendas': grupo_idade_genero.size()})
# Implemente aqui sua solução
```

## Consumidores Mais Populares (Top 5)

```python
grupo_login = purchase_file.groupby('Login')
qtde_vendas_por_login = grupo_login.size()
media_vendas_por_login = grupo_login['Valor'].mean()
rendimento_por_login = grupo_login['Valor'].sum()

df = pd.DataFrame({'Qtde vendas': qtde_vendas_por_login,
                      'Média preço': media_vendas_por_login,
                      'Rentabilidade': rendimento_por_login})
display(df.nlargest(5, 'Rentabilidade'))

grupo_itens = purchase_file.groupby('Nome do Item')

display(pd.DataFrame({'Qtde vendida': grupo_itens.size().nlargest(5)}))
# Implemente aqui sua solução
```

## Itens Mais Populares

```python
grupo_itens = purchase_file.groupby(['Nome do Item', 'Item ID', 'Valor'])

qtde_vendas_por_item = grupo_itens.size()
rendimento_por_item = grupo_itens['Valor'].sum()

pd.DataFrame({'Qtde vendas': qtde_vendas_por_item,
              'Rendimento': rendimento_por_item}).nlargest(5, 'Qtde vendas')
# Implemente aqui sua solução
```

## Itens Mais Lucrativos

```python
pd.DataFrame({'Qtde vendas': qtde_vendas_por_item,
              'Rendimento': rendimento_por_item}).nlargest(5, 'Rendimento')
# Implemente aqui sua solução
```
